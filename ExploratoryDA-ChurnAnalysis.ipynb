```python
New_Data_Dummies['MonthlyCharges'].corr(New_Data_Dummies['TotalCharges'])
# Shows that they are Positively Correlated, 
# Mean if one increases the other one also increases.
```

```python
# Churn by monthly charges and total charges
Mth = sns.kdeplot(New_Data_Dummies.MonthlyCharges[(New_Data_Dummies['Churn']==0)], color ='Red', fill = True)
Mth = sns.kdeplot(New_Data_Dummies.MonthlyCharges[(New_Data_Dummies['Churn']==1)], color ='BLue', ax = Mth , fill = True)
Mth.legend(['No Churn', 'Churn'], loc = 'upper right')
Mth.set_ylabel("Density")
Mth.set_xlabel("Monthly Charges")
Mth.set_title("Monthly Charges by Churn")
plt.show()
```

<b>Higher the Monthly Charges, Higher is the Churn Rate</b>

```python
# Churn by monthly charges and total charges
Mth = sns.kdeplot(New_Data_Dummies.TotalCharges[(New_Data_Dummies['Churn']==0)], color ='Red', fill = True)
Mth = sns.kdeplot(New_Data_Dummies.TotalCharges[(New_Data_Dummies['Churn']==1)], color ='BLue', ax = Mth , fill = True)
Mth.legend(['No Churn', 'Churn'], loc = 'upper right')
Mth.set_ylabel("Density")
Mth.set_xlabel("Total Charges")
Mth.set_title("Total Charges by Churn")
plt.show()
```

<b>Suprising Insight:</b> Higher the Churn Rate, Lower the Total Charges.
However if we combine the insights of 3 parameters i.e Tenure, Monthly Charges & Total Charges, then the picture would be more clear. 

<b>Higher Monthly Chargesat Lower Tenure results into lower total charge. Hence, all these 3 Factors viz Higher Montly Charges, Lower Tenure, and Lower Total harges are linked to Higher Churn Rate</b>

```python
def uniplot(df, col, title, hue = None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams['axes.labelsize'] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30

    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4 * len(temp.unique())
    fig.set_size_inches(width, 8)
    plt.xticks(rotation = 45)
    plt.title(title)
    ax = sns.countplot(data = df, x = col, order = df[col].value_counts().index, hue = hue, palette = 'magma')
    plt.show()
```
